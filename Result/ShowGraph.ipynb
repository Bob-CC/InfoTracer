```python
import matplotlib.pyplot as plt
import numpy as np

# 数据
cases = ['Case1', 'Case2', 'Case3', 'Case4']
x = np.arange(len(cases))

# InfoTracer数据
info_p = [1.0000, 1.0000, 1.0000, 1.0000]
info_r = [1.0000, 1.0000, 0.8750, 0.8750]

# DEPIMPACT数据
dep_p = [0.2667, 0.1064, 0.1111, 0.1159]
dep_r = [0.2500, 0.3125, 0.4375, 0.5000]

# ATLAS数据
atlas_p = [0.9412, 0.5238, 0.6190, 0.7059]
atlas_r = [1.0000, 0.6875, 0.8125, 0.7500]

# 设置图形样式
# plt.style.use('seaborn-v0_8-darkgrid')
fig, ax = plt.subplots(figsize=(14, 8))

# 颜色和标记设置
colors = {'InfoTracer': '#FF0000', 'DEPIMPACT': '#0000FF', 'ATLAS': '#008000'}
markers = {'P': 'o', 'R': 's'}

# 绘制所有线
# InfoTracer
ax.plot(cases, info_p, marker=markers['P'], markersize=8, linewidth=2, 
        color=colors['InfoTracer'], linestyle='-', label='InfoTracer-P')
ax.plot(cases, info_r, marker=markers['R'], markersize=8, linewidth=2, 
        color=colors['InfoTracer'], linestyle='--', label='InfoTracer-R')

# DEPIMPACT
ax.plot(cases, dep_p, marker=markers['P'], markersize=8, linewidth=2, 
        color=colors['DEPIMPACT'], linestyle='-', label='DEPIMPACT-P')
ax.plot(cases, dep_r, marker=markers['R'], markersize=8, linewidth=2, 
        color=colors['DEPIMPACT'], linestyle='--', label='DEPIMPACT-R')

# ATLAS
ax.plot(cases, atlas_p, marker=markers['P'], markersize=8, linewidth=2, 
        color=colors['ATLAS'], linestyle='-', label='ATLAS-P')
ax.plot(cases, atlas_r, marker=markers['R'], markersize=8, linewidth=2, 
        color=colors['ATLAS'], linestyle='--', label='ATLAS-R')

# 添加数值标签
for i in range(len(cases)):
    # InfoTracer
    ax.annotate(f'{info_p[i]:.3f}', (cases[i], info_p[i]), 
                textcoords="offset points", xytext=(0, 10), ha='center', fontsize=8, color=colors['InfoTracer'])
    ax.annotate(f'{info_r[i]:.3f}', (cases[i], info_r[i]), 
                textcoords="offset points", xytext=(0, -15), ha='center', fontsize=8, color=colors['InfoTracer'])
    
    # DEPIMPACT
    ax.annotate(f'{dep_p[i]:.3f}', (cases[i], dep_p[i]), 
                textcoords="offset points", xytext=(0, 10), ha='center', fontsize=8, color=colors['DEPIMPACT'])
    ax.annotate(f'{dep_r[i]:.3f}', (cases[i], dep_r[i]), 
                textcoords="offset points", xytext=(0, -15), ha='center', fontsize=8, color=colors['DEPIMPACT'])
    
    # ATLAS
    ax.annotate(f'{atlas_p[i]:.3f}', (cases[i], atlas_p[i]), 
                textcoords="offset points", xytext=(0, 10), ha='center', fontsize=8, color=colors['ATLAS'])
    ax.annotate(f'{atlas_r[i]:.3f}', (cases[i], atlas_r[i]), 
                textcoords="offset points", xytext=(0, -15), ha='center', fontsize=8, color=colors['ATLAS'])

# 设置图表属性
ax.set_xlabel('Case', fontsize=14, fontweight='bold')
ax.set_ylabel('Score', fontsize=14, fontweight='bold')
ax.set_title('Precision and Recall Comparison Across All Works', fontsize=16, fontweight='bold')
ax.set_ylim(0, 1.1)
ax.grid(True, alpha=0.3, linestyle='--')

# 设置图例（两列显示）
ax.legend(loc='upper left', bbox_to_anchor=(1.02, 1), ncol=1, fontsize=10)

# 添加辅助说明
fig.text(0.15, 0.02, 'Solid Line: Precision (P) | Dashed Line: Recall (R)', 
         fontsize=11, style='italic', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))
fig.text(0.15, 0.05, '○: InfoTracer | □: DEPIMPACT | △: ATLAS', 
         fontsize=11, style='italic', bbox=dict(boxstyle='round', facecolor='wheat', alpha=0.5))

plt.tight_layout()
plt.show()
```
